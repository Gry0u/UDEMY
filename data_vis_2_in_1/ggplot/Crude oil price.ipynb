```python
from ggplot import  *
import pandas as pd
```

# Data

```python
crude = pd.read_csv('data-crude-price-daily.csv')
```

```python
crude['Date'] = pd.to_datetime(crude['Date'])
crude.info()
```

```python
crude.head()
```

# Visualizations
## Price

```python
g = ggplot(aes(x='Date', y='Price'), data=crude)
g + geom_area()
```

## Price change

```python
ggplot(aes(x='Date', y='Change %'), data=crude) + geom_line()
```

## Moving average

```python
crude['avg_50d'] = crude['Price'].rolling(window=50).mean()
```

```python
ggplot(aes(x='Date'), data=crude) +\
    geom_line(aes(y='Price'), color='black') +\
    geom_line(aes(y='avg_50d'), color='red') +\
    ggtitle('Crude oil prices vs moving average') +\
    xlab('Date') +\
    ylab('USD')
    
```
