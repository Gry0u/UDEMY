# Import

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.rcParams['figure.figsize'] =(12,7)
```

# Data

```python
df = pd.read_csv('data-movies.csv', index_col='Rank')
```

```python
df.head()
```

```python
df.info()
```

# Visualizations

```python
sns.barplot(y='Title', x= 'Worldwide', data=df)
```

```python
sns.barplot(y='Title', x= 'Worldwide', data=df[:10])
```

```python
sns.barplot(y='Title', x= 'Worldwide', data=df[-10:])
```

```python
top_domestic = df.sort_values(by='Domestic', ascending=True)[:10]
top_domestic
```

```python
sns.barplot(y='Title', x='Domestic', data=top_domestic)
sns.despine(left=True)
```

## Grids

```python
g = sns.PairGrid(data=df, vars=['Worldwide', 'Domestic', 'Overseas'], hue='Studio')
g.map(plt.scatter)
g.add_legend()
```

## Custom Plots
### Viewing seaborn palettes

```python
for pal in ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']:
    sns.palplot(sns.color_palette(pal))
```

```python
sns.boxplot(data=df)
```

## matplotlib colormaps as color palettes

```python
sns.palplot(sns.color_palette('RdBu'))
sns.palplot(sns.color_palette('Blues_d'))
```

```python
sns.set_palette('Blues_d')
sns.boxplot(data=df)
```

## Building custom color palette

```python
my_palette = ['#4B0082', '#0000FF', '#00FF00', '#FF0000', '#FF7F00', '#FFFF00']
sns.set_palette(my_palette)
sns.palplot(sns.color_palette())
```

```python
sns.boxplot(data=df)
```

## Plotting context presets
Affect size of labels, lines etc but not overall style

```python
sns.set() # reset all styles to default
sns.set_context('paper') #set context
```

```python
sns.boxplot(data=df);
```

```python
sns.set()
sns.set_context('poster')
sns.boxplot(data=df);
```
