# FOREX price performance
The [foreign exchange market](https://en.wikipedia.org/wiki/Foreign_exchange_market) (Forex, FX, or currency market) is a global decentralized or over-the-counter (OTC) market for the trading of currencies.  
Here I will visualize Forex price, percentage change and 50 days moving average.

```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

## Read in Data

```python
forex_df = pd.read_csv('data-forex-usd-eur.csv')
```

## Explore Data

```python
forex_df.head()
```

```python
forex_df.info()
```

```python
forex_df['Date'] = pd.to_datetime(forex_df['Date'])
forex_df.set_index('Date', inplace=True)
```

## Visualizations
### Price

```python
plt.figure(figsize=(12,6))
forex_df['Price'].plot();
```

### Change %

```python
plt.figure(figsize=(12,6))
forex_df['Change %'].plot();
```

### 50 days moving average

```python
forex_df['avg_50d'] = forex_df['Price'].rolling(50).mean()
data = forex_df[['Price', 'avg_50d']][-365:]
data.plot(figsize=(12,5));
```

```python
data = forex_df[['Price', 'avg_50d']][-365:]
data.plot(figsize=(12,5), subplots=True);
```
