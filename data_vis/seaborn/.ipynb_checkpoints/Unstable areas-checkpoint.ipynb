```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] =(12,7)
```

```python
df = pd.read_csv('data-fragile.csv', index_col='Rank', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.info()
```

# Visualization

```python
sns.swarmplot(x='Country', y='2017 score', data=df)
```

# barplot

```python
sns.barplot(x='Country', y='2017 score', data=df[:20])
```

## Top 20 least stable countries 

```python
sns.barplot(y='Country', x='2017 score', data=df[:20])
```

# Top 20 most stable countries

```python
sns.barplot(y='Country', x='2017 score', data=df[-20:])
```

## Per continent

```python
sns.stripplot(x='Continent', y='2017 score', data=df)
```

```python
sns.boxplot(x='Continent', y='2017 score', data=df)
```

# Advanced customizations
## Styles

```python
sns.set_style('darkgrid') # will remain defaulf style we change style again
```

```python
sns.boxplot(x='Continent', y='2017 score', data=df)
```

## Palette

```python
sns.boxplot(x='Continent', y='2017 score', data=df, palette='gist_stern')
```

```python
sns.set_palette('Reds')
```

```python
sns.boxplot(x='Continent', y='2017 score', data=df)
```
