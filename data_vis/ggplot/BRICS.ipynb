# BRICS and crude oil price

## Import

```python
from ggplot import *
import pandas as pd
```

## Data

```python
gdp = pd.read_csv('data-gdp.csv')
gdp.head()
```

```python
gdp.info()
```

## Visualizations
### Brazil GDP

```python
ggplot(aes(x='Year', y='Brazil'), data=gdp) + geom_point() + geom_line()
```

### Multiple data variables on same plot

```python
g = ggplot(aes(x='Year'), data=gdp)
colors = ['black', 'red', 'blue', 'green', 'orange']
for col, color in zip(list(gdp.columns), colors):
    if col != 'Year':
        g += geom_line(aes(y=col), color=color)
        
g
```

### BRICS GDP Growth rates

```python
gdp_rates = pd.read_csv('data-gdp-rates.csv')
```

```python
gdp_rates.info()
```

```python
gdp_rates.head()
```

#### Barplot for one country

```python
ggplot(aes(x='Year', weight='Brazil'), data=gdp_rates) +\
    geom_bar()
```

#### Several countries growth rates

```python
g = ggplot(aes(x='Year'), data=gdp_rates)
colors = ['black', 'red', 'blue', 'green', 'orange']
for col, color in zip(list(gdp_rates.columns), colors):
    if col != 'Year':
        g += geom_point(aes(y=col), color=color)
        
g
```
